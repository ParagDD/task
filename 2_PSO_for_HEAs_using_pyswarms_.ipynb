```python
# X: Input features (composition, processing parameters, etc.)
# y: Target variable (yield strength)

X = df.iloc[:, 1:22]
y = df['Yield strength_Mpa']
X
```

### PSO using pyswarms 

##### Trained Random forest model first, and then applied the PSO on it.

```python
import pyswarms as ps
import numpy as np
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor()
RFR = rfr.fit(X,y)

num_elements = len(X)
def unpack_position(position):
    # Reshape the position array into a 2D array
    composition_matrix = np.reshape(position, (-1, num_elements))

    # Normalize the composition matrix if needed
    normalized_composition = composition_matrix / np.sum(composition_matrix, axis=1, keepdims=True)

    # Format the composition appropriately for your model
    # Example: Convert the normalized composition matrix into a flattened array
    formatted_composition = normalized_composition.flatten()

    return formatted_composition

# Define the objective function for yield strength prediction
def objective_function(position):
    # Convert the position into input features for prediction
    composition = unpack_position(position)

        # Perform prediction using a trained model or algorithm
    predicted_strength = RFR.predict([composition])
    # Calculate the error between predicted and actual strength
    error = mean_squared_error(y, predicted_strength)

    # Since PSO aims to minimize the objective function, return the error
    return error

# Set the number of particles and iterations

num_iterations = 10
dimensions = len(X)  # Number of elements in the alloy

bounds = [(0, 1)] * dimensions  # Example bounds for composition values

# Set the options for the PSO optimizer
options = {'c1': 0.5, 'c2': 0.3, 'w': 0.9}

# Create the PSO optimizer
optimizer = ps.single.GlobalBestPSO(n_particles=21, dimensions= 2, options=options, bounds=bounds)

# Perform the optimization
best_position, best_fitness = optimizer.optimize(objective_function, iters=num_iterations)

# Obtain the final predicted yield strength
predicted_yield_strength = objective_function(best_position)

print("Final Predicted Yield Strength:", predicted_yield_strength)
print("Best Composition:", best_position)
```
